```
import csv

from spellchecker import SpellChecker
spell = SpellChecker()
```

```

def removeSpace(string):
  if (string[0] == " "):
    return string[1:]
  else:
    return string

def removeNewLine(string):
  if string[len(string)-1] == '\n':
    return string[0:len(string)-1]
  else:
    return string


f = open("/content/drive/My Drive/MLResearchProject/Data_Sets/Raw_Format/Questions.txt", "r")
lines = f.readlines()
print("Test")

columns = ['Question', 'Answer 1', 'Answer 2', 'Answer 3', 'Answer 4', 'Answer 5', 'Explanation', 'Author']
rows = []

#Iterate through all lines
for i in range (0, len(lines)):
  print(i)
  words = spell.split_words(lines[i])
  [spell.correction(word) for word in words]
  if (len(lines[i]) > 2):
    if (lines[i][0:2] == "ID"):
      i += 2
      singleRow = []
      singleRow.append(removeNewLine(removeSpace(lines[i])))
      i = i + 1
      if (lines[i][0:1] == "*"):
        singleRow.append(removeNewLine(removeSpace(lines[i][3:])))
        i = i + 1
      else:
        singleRow.append(removeNewLine(removeSpace(lines[i][2:])))
        i = i + 1
      if (lines[i][0:1] == "*"):
        singleRow.append(removeNewLine(removeSpace(lines[i][3:])))
        i = i + 1
      else:
        singleRow.append(removeNewLine(removeSpace(lines[i][2:])))
        i = i + 1
      if (lines[i][0:1] == "*"):
        singleRow.append(removeNewLine(removeSpace(lines[i][3:])))
        i = i + 1
      else:
        singleRow.append(removeNewLine(removeSpace(lines[i][2:])))
        i = i + 1
      if (lines[i][0:1] == "*"):
        singleRow.append(removeNewLine(removeSpace(lines[i][3:])))
        i = i + 1
      else:
        singleRow.append(removeNewLine(removeSpace(lines[i][2:])))
        i = i + 1
      if (lines[i][0:2] == "e)"):
        singleRow.append(removeNewLine(removeSpace(lines[i][2:])))
        i = i + 1
      elif (lines[i][0:3] == "*e)"):
        singleRow.append(removeNewLine(removeSpace(lines[i][3:])))
      else:
        singleRow.append("")
      if (lines[i][0:11] == "Explanation"):
        singleRow.append(lines[i][13:])
        rows.append(singleRow)
        i = i + 1
      else:
        singleRow.append("")
      if (lines[i][0:6] == "Author"):
        singleRow.append(lines[i][8:])
        rows.append(singleRow)
        i = i + 1
      else:
        singleRow.append("")

with open('Questions.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(columns)
    for i in range (0, len(rows)):
      writer.writerow(rows[i])
```
