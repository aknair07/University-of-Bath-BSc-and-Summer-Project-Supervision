```
rows = data.shape[0]
columns = data.shape[1]

data.shape
```

```
new_data = pd.DataFrame(data=None, columns=['Optimal Topic Number'])
new_data.head()
```

```
import numpy as np

for column in range(0,columns):
  for row in range(0,rows):
    if (np.isnan(data.iloc[row][column]) == True):
      x = data.iloc[row-1][column]
      break
    elif (row == rows-1):
      x = data.iloc[row][column]
      break
  new_data.loc[column] = x
```
